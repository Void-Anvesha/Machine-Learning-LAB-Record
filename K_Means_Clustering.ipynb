```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 2], [1, 4], [1, 0],
              [10, 2], [10, 4], [10, 0]])
```

```python
  # Number of clusters
k=2
```

```python
#Initializing centroid by selecting 1st 2 points
centroids=X[:k]
```

```python
for _ in range(5):
  clusters=[]
  for point in X:
    #Calculate the Euclidean distance from the current point to each centroid.
     distances = np.linalg.norm(point - centroids, axis=1)
     clusters.append(np.argmin(distances))
  clusters = np.array(clusters)


 # Step 4: Update centroids as mean of points in each cluster
  for i in range(k):
        points_in_cluster = X[clusters == i]
        centroids[i] = points_in_cluster.mean(axis=0)
# Output results
print("Final centroids:\n", centroids)
print("Cluster assignments:", clusters)



# Plot points and centroids
colors = ['r', 'b']
for i in range(k):
    points = X[clusters == i]
    plt.scatter(points[:, 0], points[:, 1], c=colors[i])
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='y', marker='X')  # centroids
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate a small sample dataset
np.random.seed(42)
X1 = np.random.randn(10, 2) + np.array([2, 2])
X2 = np.random.randn(10, 2) + np.array([-2, -2])
X3 = np.random.randn(10, 2) + np.array([2, -2])
X = np.vstack([X1, X2, X3])

# BEFORE clustering plot
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c='gray', s=50)
plt.title('Before Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Apply K-means
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# AFTER clustering plot
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('After K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('FeatureÂ 2')
plt.show()
```
